```
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK stopwords
nltk.download('stopwords')
nltk.download('punkt')

# Define the medicine dataset as a list of dictionaries
medicine_data = [
    {'ID': 1, 'Name': 'Paracetamol', 'Type': 'Tablet', 'Dosage': '500mg', 'Price': 5.99, 'Quantity': 100, 'Review': 'Effective for reducing fever and mild to moderate pain.'},
    {'ID': 2, 'Name': 'Aspirin', 'Type': 'Capsule', 'Dosage': '300mg', 'Price': 3.49, 'Quantity': 80, 'Review': 'Commonly used to relieve mild to moderate pain and reduce fever.'},
    {'ID': 3, 'Name': 'Ibuprofen', 'Type': 'Tablet', 'Dosage': '200mg', 'Price': 7.25, 'Quantity': 120, 'Review': 'A nonsteroidal anti-inflammatory drug (NSAID) used to relieve pain, reduce inflammation, and lower fever.'},
    {'ID': 4, 'Name': 'Omeprazole', 'Type': 'Capsule', 'Dosage': '20mg', 'Price': 9.99, 'Quantity': 60, 'Review': 'Used to treat certain stomach and esophagus problems (such as acid reflux, ulcers).'},
    {'ID': 5, 'Name': 'Cetirizine', 'Type': 'Tablet', 'Dosage': '10mg', 'Price': 6.75, 'Quantity': 90, 'Review': 'An antihistamine used to relieve allergy symptoms such as watery eyes, runny nose, itching eyes/nose, and sneezing.'},
    {'ID': 6, 'Name': 'Loratadine', 'Type': 'Tablet', 'Dosage': '10mg', 'Price': 8.49, 'Quantity': 70, 'Review': 'An antihistamine used to relieve allergy symptoms such as watery eyes, runny nose, itching eyes/nose, and sneezing.'},
    {'ID': 7, 'Name': 'Simvastatin', 'Type': 'Tablet', 'Dosage': '20mg', 'Price': 12.99, 'Quantity': 50, 'Review': 'Used along with a proper diet to help lower "bad" cholesterol and fats (such as LDL, triglycerides) and raise "good" cholesterol (HDL) in the blood.'},
    {'ID': 8, 'Name': 'Metformin', 'Type': 'Tablet', 'Dosage': '500mg', 'Price': 11.75, 'Quantity': 40, 'Review': 'An oral diabetes medicine that helps control blood sugar levels.'},
    {'ID': 9, 'Name': 'Warfarin', 'Type': 'Tablet', 'Dosage': '5mg', 'Price': 14.25, 'Quantity': 30, 'Review': 'Used to treat or prevent blood clots, including deep vein thrombosis and pulmonary embolism.'},
    {'ID': 10, 'Name': 'Fluoxetine', 'Type': 'Capsule', 'Dosage': '20mg', 'Price': 10.99, 'Quantity': 50, 'Review': 'An antidepressant used to treat depression, panic attacks, obsessive-compulsive disorder (OCD), bulimia nervosa, and premenstrual dysphoric disorder (PMDD).'}
]

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(medicine_data)

# Define stop words
stop_words = set(stopwords.words('english'))

# Define function to remove stop words
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_text = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

# Define function for word frequency analysis
def word_frequency(text):
    tokens = word_tokenize(text)
    freq_dist = nltk.FreqDist(tokens)
    return freq_dist

# Define function to generate word cloud
def generate_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Define pipeline
def text_processing_pipeline(text):
    text = remove_stopwords(text)
    freq_dist = word_frequency(text)
    generate_wordcloud(text)
    return freq_dist

# Apply pipeline to medicine reviews
for index, row in df.iterrows():
    print("Medicine:", row['Name'])
    print("Review:", row['Review'])
    freq_dist = text_processing_pipeline(row['Review'])
    print("Word Frequencies:", freq_dist)
    print("="*50)
```

```
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')

# Define the health-related dataset
health_dataset = [
    "I recently started a new fitness routine and I feel great! My energy levels are up and I'm sleeping better.",
    "I've been struggling with chronic pain for years, and it's really affecting my quality of life. I'm desperate for relief.",
    "I just received some concerning test results from my doctor, and I'm feeling really anxious about what it could mean for my health.",
    "I've been trying to eat healthier and incorporate more fruits and vegetables into my diet, but it's been a challenge to stay consistent.",
    "I had a check-up with my doctor today, and I'm relieved to hear that my blood pressure and cholesterol levels are within the normal range.",
    "I've been feeling really fatigued lately, and I'm worried that it could be a sign of a more serious health issue.",
    "I'm proud of myself for quitting smoking and making positive changes to improve my lung health.",
    "I injured myself while exercising, and now I'm worried that I won't be able to continue with my fitness routine.",
    "I've been taking medication for my condition, and I'm happy to report that I'm seeing some improvement in my symptoms.",
    "I'm feeling optimistic about my health goals for the new year and excited to make progress towards them."
]

# Initialize the sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each sentence in the health-related dataset
for i, sentence in enumerate(health_dataset):
    # Analyze the sentiment of the sentence
    sentiment_scores = sid.polarity_scores(sentence)

    # Determine the sentiment label based on the compound score
    if sentiment_scores['compound'] >= 0.05:
        sentiment_label = 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    # Print the sentiment analysis results
    print(f"Sentence {i+1}: {sentence}")
    print(f"Sentiment Score: {sentiment_scores['compound']}")
    print(f"Sentiment Label: {sentiment_label}")
    print("=" * 50)
```

```
# Define health-related dataset
health_reviews = [
    "I feel great after starting this new exercise routine. My energy levels are up, and I'm sleeping better at night.",
    "I've been experiencing some pain in my lower back lately. It's been affecting my daily activities and I'm concerned.",
    "The medication prescribed by my doctor has been very effective in managing my symptoms. I'm feeling much better now.",
    "I'm struggling with maintaining a healthy diet. I often give in to unhealthy cravings and feel guilty afterwards.",
    "The side effects of this treatment are unbearable. I'm experiencing nausea, headaches, and fatigue on a daily basis.",
    "I've been following this wellness program for a few weeks now, and I'm already seeing positive changes in my mood and overall well-being.",
    "I'm feeling stressed and overwhelmed with work. It's taking a toll on my mental and physical health.",
    "This new supplement has made a significant difference in my joint pain. I can now move more freely without discomfort.",
    "I'm disappointed with the lack of progress in my fitness journey. Despite my efforts, I haven't seen any improvements in my strength or endurance.",
    "I'm grateful for the support of my healthcare team. They've been instrumental in helping me manage my chronic condition."
]

# Define positive and negative word lists from Bing Liu lexicon
positive_words = set([
    "great", "effective", "better", "positive", "wellness", "significant", "grateful"
])

negative_words = set([
    "pain", "struggling", "unbearable", "stress", "overwhelmed", "disappointed", "guilty"
])

# Define function to calculate Bing Liu sentiment score for a given review
def bing_liu_sentiment(review):
    # Tokenize the review into words
    words = review.lower().split()

    # Count occurrences of positive and negative words
    positive_count = sum(1 for word in words if word in positive_words)
    negative_count = sum(1 for word in words if word in negative_words)

    # Calculate sentiment score based on word counts
    score = positive_count - negative_count

    return score

# Perform sentiment analysis using Bing Liu lexicon on each health review
for i, review in enumerate(health_reviews):
    # Calculate Bing Liu sentiment score for the review
    score = bing_liu_sentiment(review)

    # Determine sentiment label based on score
    if score > 0:
        sentiment_label = 'Positive'
    elif score < 0:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    # Print sentiment analysis results
    print(f"Review {i+1}: {review}")
    print(f"Bing Liu Sentiment Score: {score}")
    print(f"Sentiment Label: {sentiment_label}")
    print("=" * 50)
```
